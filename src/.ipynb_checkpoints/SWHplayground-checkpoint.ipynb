```python
import json
import requests
```

```python
def jprint(obj):
    # create a formatted string of the Python JSON object
    print(json.dumps(obj, sort_keys=True, indent=4))
```

```python
resp = requests.get("https://archive.softwareheritage.org/api/1/stat/counters/")
counters = resp.json()
jprint(counters)
```

```python
resp = requests.get("https://archive.softwareheritage.org/api/1/origin/search/narps-open/")
origins = resp.json()
print(f"We found {len(origins)} entries.")
for origin in origins[1:10]:
    print(f"- {origin['url']}")
```

```python
resp = requests.get("https://archive.softwareheritage.org/api/1/origin/https://github.com/Inria-Empenn/narps_open_pipelines/get/")
found = resp.json()
jprint(found)
```

```python
resp = requests.get("https://archive.softwareheritage.org/api/1/origin/https://github.com/Inria-Empenn/narps_open_pipelines/visits/")
found = resp.json()
length = len(found)
print(f"Number of visits found: {format(length)}.")
print("With dates:")
for visit in found:
    print(f"- {visit['visit']} {visit['date']}")
print("\nExample of a single visit entry:")
jprint(found[0])
```

```python
# Store snapshot id
snapshot = found[0]['snapshot']
print(f"Snapshot is {format(snapshot)}.")
```

```python
snapshotr = requests.get("https://archive.softwareheritage.org/api/1/snapshot/{}/".format(snapshot))
snapshotj = snapshotr.json()
jprint(snapshotj)
```

```python
print(f"Revision ID is {snapshotj['id']}.")
master_url = snapshotj['branches']['refs/heads/main']['target_url']
masterr = requests.get(master_url)
masterj = masterr.json()
jprint(masterj)
```

```python
root_url = masterj['directory_url']
rootr = requests.get(root_url)
rootj = rootr.json()
for f in rootj:
    print(f"- {f['name']}.")
```
